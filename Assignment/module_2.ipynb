### How memory is managed in Python?

##### In python we do not need to worry about memory. Python has garbage collector. This is automatic process that allocate and deallocate memory. For instance, there is one object created "a" and later on that in memory no reference is pointing to this object so, virtual machine has garbage collector and this garbage collector deletes all the object which is not in use.

Python works with two type of memory : 

1.   Stack Memory
2.   Heap Memory

---

1. Stack Memory :
>  In this Stack memory allocation happens in contiguous block of memory.
Methods , methods calls and reference are stored in stack memory. When a function is called all the variables are allocate memory in Stack for temporarily on function call Stack and once function returns all variables are deleted from memory. All this allocation and deallocation happends on contiguous memeory. 
```
Example :
def funct() :
      a = 20
      b = []
      c = ""
```
Here inside this funct() all the variables are stored in Stack memory. 
---

2. Heap Memory :
>  The memory is allocated during the execution of instructions written by programmers. The variables are needed outside of function or are shared with multiple functions globally are stored in Heap memory.
```
Example :
num = [-23,22,11,42]
def funct() :
      a = 20
def funct2():
      b = 30
```
Here num is stored in Heap memory.

### What is the purpose continue statement in python?

##### Continue statement is a loop control statement that forces to execute the next iteration of the loop while skipping the rest of the code inside the loop for the current iteration only i.e. when the continue statement is executed in the loop, the code inside the loop following the continue statement will be skipped for the current iteration and the next iteration of the loop will begin.

```
Example : 

for i in range(1,11):
  if i==4:
    continue
  print(i)
  i+=1
```
In the above example, we placed continue statement on particular condition.
If i == 4 then this loop will skip this part and the next iteration of the loop will begin. Loop will not print 4 it will skip it.

### Write a Python function to reverses a string if its length is a multiple of 4.

````python
def rev_string(strr):
  if len(strr) % 4 == 0:
    return "".join(reversed(strr))

strr = input("Enter any string : ")
print("String :",rev_string(strr))
````

### Write a Python program to get a string made of the first 2 and the last 2 chars from a given a string. If the string length is less than 2,return instead of the empty string.

````python
def make_str(string):
  str_len = len(string) 
  if str_len < 2:
    return string
  else:
    return string[:2] + string[-2:]

strr = input("Enter any string : ")
print(make_str(strr))
````

### Write a Python function to insert a string in the middle of a string.

````python
def add_mid_str(string,mid_string):
  str_len = len(string)//2
  return string[:str_len] + mid_string + string[str_len:]

str1 = input("Enter any string : ")
str2 = input("Enter string to add in middle : ")

print("String : ",add_mid_str(str1,str2))
````
