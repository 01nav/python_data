# Class Method
#### Class method is used to modify class variables.

```python
class Car:
  wheels = 4

  def driving(self):
    print("Himal Mevada is driving a car.")

  @classmethod
  def change_wheels(cls,val):
    cls.wheels = val
  
```

```python
obj_car = Car()
obj_car.driving()

obj_car.wheels = 10
print(obj_car.wheels)

print(Car.wheels)

# Here i can change wheels in car class by accessing wheels from direct class.
Car.wheels = 9
print(Car.wheels)

# What if i want to change class variable from instance of class.
# Then class method comes into picture.
obj_car.change_wheels(12)
print(Car.wheels)
```

## DECORATORS 
#### A decorator is a design pattern in Python that allows a user to add new functionality to an existing object without modifying its structure. Decorators are usually called before the definition of a function you want to decorate.

```python
def decorator1(func1):
  def dec_execute():
    print("-----------------------------------------------------------------------")
    func1()
    print("-----------------------------------------------------------------------")
  return dec_execute()

@decorator1
def himal():
  print("I want decoration before and after this line.")
```

```python
# Method Overriding --->

# Accessing member variable and member function from parent or sibling class

class Vehicle:
  c1_var1 = "red"
  def __init__(self):
    self.c1_var1 = "white"
    self.var1 = "This is vehicle class's constructor."
    self.data = "data is here"

class Mercedes(Vehicle):
  # pass
  c1_var1 = "magento"
  def __init__(self):
    # If i want to call parent class's constructor then we need to use super() method.
    # This method is used to access parent class or sibling class's data member and member methods.
    
    self.c1_var1 = "green"
    self.var1 = "This is Mercedes class's constructor."
    super().__init__() # Constructor.
    print(super().c1_var1) # Class variable
```

### Operator Overloading -------->
#### Operator Overloading means giving extended meaning beyond their predefined operational meaning.
#### For example operator + is used to add two integers as well as join two strings and merge two lists. 
#### It is achievable because ‘+’ operator is overloaded by int class and str class. 
#### You might have noticed that the same built-in operator or function shows different 
#### behavior for objects of different classes, this is called Operator Overloading. 

```python
class Emp:

  def __init__(self,name,year,work):
    self.name = name
    self.year = year
    self.work = work

  def details(self):
    print(f"{self.name} has {self.year} experience in {self.work} field.")

  # We have overwirtten the "+" operator using dunder or magic method.
  def __add__(self,other):
    return self.name + other.name

  # The __repr__() function returns a printable representation of the given object.
  def __repr__(self):
    return f"{self.name} has {self.year} experience in {self.work} field."
  
  # The __str__() method represents the class objects as a string
  def __str__(self):
    return f"Emp('{self.name}', {self.year}, '{self.work}')"
```

```python
ob1 = Emp("Himal",2,"Programming")
ob2 = Emp("Jigar",3,"Civil Engineering")
ob1.details()
ob2.details()
```

```python
print(ob1 + ob2)
```

```python
# We can also call this method by using there name like stpr() and str().
# When we write str() if str() is not found but repr() is there then it will execute repr() function.
print(str(ob2))
print(repr(ob2))
```

### Abstract Class and Abstract Method :
#### A class which contains one or more abstract methods is called an abstract class.
#### An abstract class can be considered as a blueprint for other classes. It allows you to create a set of methods that must be created within any child classes built from the abstract class.
#### While we are designing large functional units we use an abstract class. When we want to provide a common interface for different implementations of a component, we use an abstract class. 

```python
# When we want to set some sort of common functionality or implementation in our programm we use an 
# abstract class.
from abc import ABC,abstractmethod

class Impliment(ABC):

  @abstractmethod
  def pr_details():
    return 0

class Emp(Impliment):

  def __init__(self,name,year,work):
    self.name = name
    self.year = year
    self.work = work

  def pr_details(self):
    print(f"{self.name} has {self.year} experience in {self.work} field.")

  def __repr__(self):
    return f"Emp('{self.name}',{self.year},'{self.work}')"
```

```python
emp = Emp("himal",4,"Programmer")
```

```python
emp.pr_details()
```

```python
print(repr(emp))
```

## Setters and Property Decorators
#### When we want to define property of we use property decorators. That will return attribute of class.
#### Getters and Setters play an important role in retrieving and updating the value of a variable outside the encapsulating class.
#### If we want to use getter, setter, deleter then we must create property decorator so that will return attribute and all this getter,setter and deleter works with this attribute.

```python
class Emp():

  def __init__(self,fname,lname):
    self.fname = fname
    self.lname = lname
    # self.email = f"{self.fname}.{self.lname}@gmail.com"

  @property
  def email(self):
    if self.fname == None or self.lname == None:
      return "Email is not set. Please set email using setter."
    return f"{self.fname}.{self.lname}@gmail.com"

  @email.setter
  def email(self,string):
    name = string.split("@")[0]
    self.fname = name.split(".")[0]
    self.lname = name.split(".")[1]
    self.email = f"{self.fname}.{self.lname}@gmail.com"

  @email.deleter
  def email(self):
    self.fname = None
    self.lname = None

  def pr_details(self):
    return f"Hello, I am {self.fname} {self.lname}."

  # def __repr__(self):
  #   return f"Emp('{self.fname}',{self.lname},'{self.email}')"
```
