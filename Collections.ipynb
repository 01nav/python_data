```python
set1 = {'Mevada', 'himal', 'mevada'}
set2 = {'1', '2', 'Himal', 'Ishan', 'Ishan_tup'}
```

```python
set1.union(set2) # union will add set2 item into set1. 
```

```python
set1
```

```python
set2
```

```python
set1.add('Himal')
set1.add("1")
```

```python
set1
```

```python
# intersection return a set that contains the items that exist in both set 1, and set 2:.
set3 = set1.intersection(set2) 
```

```python
set3
```

```python
# Intersection_update will removes the items that are not exist in both set.
set1.intersection_update(set2)
```

```python
print(set1,set2)
set1.add('Gujarati')
```

```python
# Difference will return new set that contains items that are only exist in set1 not in set2.
set3 = set1.difference(set2)
```

```python
set3
```

```python
# Difference_update will remove the item that exist in both set.
set1.difference_update(set2)
```

```python
set1
```

```python
set1,set2,set3
# set1.add("1")
# set1.add("2")
```

```python
# Symmetric_difference will return a new set that contains all items from both sets except item which are present in both set.
set3 = set1.symmetric_difference(set2)
```

```python
set1,set2,set3
```

```python
# Symmetric_difference_update will remove item that are present in both sets AND insert the items that is not present in both sets.
set1.symmetric_difference_update(set2)
```

```python
set1
```

```python
# It will return True if no items in set1 are present in set2. "True "both set are distinct.
set1.add("Jigar")
set1.isdisjoint(set2)
```

```python
# It will return True if all items in set2 are present in set3.
set2.issubset(set3)
```

```python
# It will return True if all items in set2 are present in set3.
set3.issuperset(set2)
```
