```python
import datetime
try:
    class Bank():

        acc_no_preset = 1000
        bank_data = dict()
        
        def create_account(self):

            acc_no = self.acc_no_preset + len(self.bank_data) + 1
            return acc_no
        
        def acc_info(self,acc_number):
            print(" Account information ".center(60, "-"), end="\n\n")
            print(f"Your account number : XXXXXXX{str(acc_number)[7:]}")
            for k, v in self.bank_data[acc_number].items():
                if k == 'transactions':
                    print(f"Your total {k} is {len(v)}")
                else:
                    print(f"Your account {k[4:]} is {v}")


        def verify_account(self,acc_number,acc_pin):
            ac_found = ""
            pin_found = ""

            for key in self.bank_data:
                if acc_number == key:
                    ac_found = True

            assert ac_found,"Account doesnot exists."

            if acc_pin == self.bank_data[acc_number]['acc_pin']:
                pin_found = True

            assert pin_found, "Wrong pin."
            
            self.acc_info(acc_number)

        def credit_money(self,acc_number):

            print(" Credit money ".center(60, "-"), end="\n\n")

            amount = input("Enter amount (without decimal value) : ")
            assert amount.isnumeric(), "Wrong input"
            amount = float(amount)
            
            self.data = {
              "amount" : amount,
              "tr_type" : 'cr',
              "datetime": datetime.datetime.now()
            }
        
            self.bank_data[acc_number]["transactions"].append(self.data)
            self.bank_data[acc_number]["acc_balance"] += amount

            print(f"{amount} credited into your account.")
            print(f"Account balance : {self.bank_data[acc_number]['acc_balance']}")


        def debit_money(self,acc_number):
            
            print(" Debit money ".center(60, "-"), end="\n\n")

            amount = input("Enter amount (without decimal value) : ")
            assert amount.isnumeric(), "Wrong input"
            amount = float(amount)
            rem_amount = self.bank_data[acc_number]['acc_balance'] - amount
            assert rem_amount > 0, "No sufficient balance."

            self.data = {
              "amount" : amount,
              "tr_type" : 'cr',
              "datetime": datetime.datetime.now()
            }
        
            self.bank_data[acc_number]["transactions"].append(self.data)
            self.bank_data[acc_number]["acc_balance"] -= amount
            
            print(f"{amount} debited from your account.")
            print(f"Account balance : {self.bank_data[acc_number]['acc_balance']}")


        def all_transactions(self,acc_number):

            trans_data = [] 
            
            decorate = [" All transactions ".center(60, "-") + "\n\n", "".center(50, "_") + "\n", "amount".center(15) + "cr/db".center(15) + "date & time".center(15)+ "\n", "".center(50, "_")]
            print(*decorate)
            
            for tran in self.bank_data[acc_number]['transactions']:
                trans_data += "\n"
                for k,v in tran.items():
                    if k == 'datetime':
                        print(f"{v.strftime('%x %X')}".center(15), end="")
                        trans_data.append(f"{v.strftime('%x %X')}".center(15))
                    else:
                        print(str(v).center(15), end="")
                        trans_data.append(f"{v}".center(15))
                print("")
   

            f = open(f"acc {acc_number}.txt", "w+")
            data = decorate + trans_data
            f.writelines(data)
            f.close()

            # TEXT FILE FORMAT -->                
            # f = open(f"acc {acc_number}.txt", "w+")
            # acc_num = f"Account Number :  {acc_number}\n"
            # acc_name = f"Account Name :  {self.bank_data[acc_number].get('acc_name')}\n"
            # acc_pin = f"Account Name :  {self.bank_data[acc_number].get('acc_pin')}\n"
            # acc_trans = f"Money Transactions :  {self.bank_data[acc_number].get('transactions')}\n"
            # data = acc_num + acc_name + acc_pin + acc_trans
            # f.write(data)
            # f.close()

            # CSV FILE FORMAT -->
            # f = open(account.csv, "w+")
            # acc_num = f"Account Number :  {acc_number}\t"
            # acc_name = f"Account Name :  {self.bank_data[acc_number].get('acc_name')}\t"
            # acc_pin = f"Account Name :  {self.bank_data[acc_number].get('acc_pin')}\t"
            # acc_trans = f"Money Transactions :  {self.bank_data[acc_number].get('transactions')}\n"
            # data = acc_num + acc_name + acc_pin + acc_trans
            # f.write(data)
            # f.close()

                
#------------------------------------------------------------------------------------------
                
    
except Exception as err:
    print(err)
```

```python
try:

    def services(acc_number):

        service = Bank()

        while True:
            print("")
            print("1 -> Account Information")
            print("2 -> Credit money")
            print("3 -> Debit money")
            print("4 -> Transactions")
            print("5 -> Exit")

            choice = input("\nEnter number to access service :")
            assert choice.isnumeric(), "Wrong input."
            choice = int(choice)

            print("")

            if choice == 1:
                service.acc_info(acc_number)

            elif choice == 2:
                service.credit_money(acc_number)

            elif choice == 3:
                service.debit_money(acc_number)

            elif choice == 4:
                service.all_transactions(acc_number)

            elif choice == 5:

                print("Thanks for contact us".center(60, "_"))
                raise Exception("")

            else:
                print("Wrong choice.")

    print('Press "1" for create new account\nPress "2" for open existing account.')
    ans = input("Enter 1 or 2 : ")
    assert ans.isnumeric(), 'Wrong input.'
    ans = int(ans)

    if ans == 1:

        acc = Bank()
        acc_no = acc.create_account()
        print(f"Your account number : {acc_no}")

        acc_name = input('Enter your name : ')
        acc_pin = input("Enter account pin : ")

        while not acc_pin.isnumeric() or not len(acc_pin) == 4:
            print("\nPlease enter four numbers only.")
            acc_pin = input("Enter account pin : ")

        acc_pin = int(acc_pin)

        bank_acc_info = {
            "acc_name": acc_name,
            "acc_pin": acc_pin,
            "acc_balance": 0,
            "transactions": []
        }
        acc.bank_data.setdefault(acc_no, bank_acc_info)
        services(acc_no)

    elif ans == 2:

        find = Bank()

        acc_no = input("Enter account number : ")
        acc_pin = input("Enter account pin : ")
        assert acc_pin.isnumeric(), "Wrong input."
        acc_no = int(acc_no)
        acc_pin = int(acc_pin)
        find.verify_account(acc_no, acc_pin)

        services(acc_no)

    else:
        print("Enter 1 or 2 only numbers.")

except Exception as err:
    print(err)
```
